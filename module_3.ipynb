```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("imdb_top_1000.csv")
df.head(2)
```

```python
df = df.drop(['Poster_Link', 'Released_Year', 'Certificate', "Runtime","Genre",
              "IMDB_Rating", "Meta_score", "Director", "Star1","Star2","Star3",
              "Star4", "No_of_Votes","Gross" ], axis=1)


```

```python
df.set_index(['Series_Title'], inplace=True)
df.head()
```

```python
#no null values in plot overiew

df.isnull().sum()
```

```python
#chatgpt assistance
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Overview'])
tfidf_matrix.toarray()
```

```python
#chatgpt assistance
target_movie = "Forrest Gump"
query_index = df.index.get_loc(target_movie)
cosine_similarities = cosine_similarity(tfidf_matrix, tfidf_matrix[query_index])[:,0]
cosine_pairs = [(movie,d) for movie, d in zip(df.index, cosine_similarities) if movie != target_movie]
cosine_top10 = sorted(cosine_pairs, key = lambda x: x[1], reverse=True)[:10]

print("Top 10 most similar movies to Forest Gump using Cosine similarity")
for movie,d in cosine_top10:
  print("\n",movie, f"-->{d:.4f}")
```
